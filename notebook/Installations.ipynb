```python
import os
import numpy as np
import pandas as pd
import matplotlib
import sklearn
import sklearn.linear_model

# Utility function
def print_package_info(pkg, name=None):
    name = name or pkg.__name__
    version = getattr(pkg, "__version__", "N/A")
    location = os.path.dirname(pkg.__file__)
    print(f"{name} version: {version}\nLocation: {location}\n")

# Examples
print_package_info(np, "numpy")
print_package_info(pd, "pandas")
print_package_info(matplotlib, "matplotlib")
print_package_info(sklearn, "scikit-learn")
print_package_info(sklearn.linear_model, "sklearn.linear_model")
```
